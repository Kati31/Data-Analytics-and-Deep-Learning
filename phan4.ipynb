# 2. Hãy tạo biến phân loại (phanlopt1) cho môn toán (T1) như sau:
- a. Từ 0 đến dưới 5 = kém (ký hiệu “k”)
- b. Từ 5 đến dưới 7 = trung bình (ký hiệu “tb”)
- c. Từ 7 đến dưới 8 = khá (ký hiệu “k”)
- d. Từ 8 trở lên = giỏi (ký hiệu “g”)

```python
import pandas as pd

file_path = "processed_dulieuxettuyendaihoc.csv"
df = pd.read_csv(file_path, encoding="utf-8")

if "T1" in df.columns:
    def classify_t1(score):
        if score < 5:
            return "k"
        elif 5 <= score < 7:
            return "tb"
        elif 7 <= score < 8:
            return "k"
        else:
            return "g"

    df["phanlopt1"] = df["T1"].apply(classify_t1)

    print(df[["T1", "phanlopt1"]].head(10))
else:
    print("Cột 'T1' không tồn tại trong dữ liệu.")
```

# 3. Lập bảng tần số cho biến phanloait1

# 4. Vẽ biểu đồ đường Multiple Line cho biến T1 được phân loại bởi biến phanlopt1

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "processed_dulieuxettuyendaihoc.csv"
df = pd.read_csv(file_path, encoding="utf-8")

df["T1_round"] = df["T1"].round(1)

t1_counts = df["T1_round"].value_counts().sort_index()

t1_values = [2.4, 2.6, 2.8, 3.4, 3.7, 4.1, 4.3, 4.5, 4.7, 4.9, 
             5.1, 5.3, 5.5, 5.7, 5.9, 6.1, 6.3, 6.5, 6.7, 7.0, 
             7.2, 7.5, 7.8, 8.1, 8.3, 8.6, 9.3]

y_counts = [t1_counts.get(val, 0) for val in t1_values]

def classify_T1(score):
    if score < 5:
        return "kém"
    elif score < 7:
        return "trung bình"
    elif score < 8:
        return "khá"
    else:
        return "giỏi"

color_map = {
    "giỏi": "blue",
    "trung bình": "purple",
    "khá": "yellow",
    "kém": "green"
}


marker_colors = [color_map[classify_T1(val)] for val in t1_values]

plt.figure(figsize=(12, 6))
plt.plot(t1_values, y_counts, color="black", linestyle="-", linewidth=1, label="Count")
plt.scatter(t1_values, y_counts, color=marker_colors, s=80, zorder=5)

plt.xticks(t1_values, rotation=45)
plt.yticks(range(1, 6))

plt.xlabel("Giá trị T1")
plt.ylabel("Count")
plt.title("Biểu đồ đường cho biến T1 với phân loại theo phanlopt1")
legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Giỏi', markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Trung bình', markerfacecolor='purple', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Khá', markerfacecolor='yellow', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Kém', markerfacecolor='green', markersize=10)
]
plt.legend(handles=legend_elements, title="Phân loại T1")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

# 5. Vẽ biểu đồ Drop-line cho biến T1 được phân loại bởi biến phanlopt1

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

file_path = "processed_dulieuxettuyendaihoc.csv"
df = pd.read_csv(file_path, encoding="utf-8")

df["T1_round"] = df["T1"].round(1)

t1_counts = df["T1_round"].value_counts().sort_index()

t1_values = [2.4, 2.6, 2.8, 3.4, 3.7, 4.1, 4.3, 4.5, 4.7, 4.9, 
             5.1, 5.3, 5.5, 5.7, 5.9, 6.1, 6.3, 6.5, 6.7, 7.0, 
             7.2, 7.5, 7.8, 8.1, 8.3, 8.6, 9.3]

y_counts = [t1_counts.get(val, 0) for val in t1_values]

def classify_T1(score):
    if score < 5:
        return "kém"
    elif score < 7:
        return "trung bình"
    elif score < 8:
        return "khá"
    else:
        return "giỏi"

color_map = {
    "giỏi": "blue",
    "trung bình": "purple",
    "khá": "yellow",
    "kém": "green"
}

marker_colors = [color_map[classify_T1(val)] for val in t1_values]

plt.figure(figsize=(12, 6))

for x, y, color in zip(t1_values, y_counts, marker_colors):
    plt.vlines(x, 0, y, colors=color, linestyles="dotted", linewidth=2)
    plt.scatter(x, y, color=color, s=100)

plt.xticks(t1_values, rotation=45)
plt.xlabel("Giá trị T1")
plt.ylabel("Count")
plt.title("Biểu đồ Drop-line của T1 với phân loại theo phanlopt1")
plt.ylim(0, max(y_counts) + 1)

legend_elements = [
    Line2D([0], [0], marker='o', color='w', label='Giỏi', markerfacecolor='blue', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Trung bình', markerfacecolor='purple', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Khá', markerfacecolor='yellow', markersize=10),
    Line2D([0], [0], marker='o', color='w', label='Kém', markerfacecolor='green', markersize=10)
]
plt.legend(handles=legend_elements, title="Phân loại T1")

plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```
