# 1. Hãy trình bày dữ liệu biến: GT
Gợi ý
- Lập bảng tần số và tần suất
- Vẽ biểu đồ tần số (cột), biểu đồ tần suất (tròn).


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "processed_dulieuxettuyendaihoc.csv"
try:
    df = pd.read_csv(file_path)
    print("Dữ liệu đã được tải thành công!")
except FileNotFoundError:
    print(f"Tệp '{file_path}' không tìm thấy. Vui lòng kiểm tra đường dẫn.")
    exit()

frequency_table = df["GT"].value_counts()
relative_frequency_table = df["GT"].value_counts(normalize=True) * 100

print("Bảng tần số của biến GT:")
print(frequency_table)
print("\nBảng tần suất của biến GT (%):")
print(relative_frequency_table)

plt.figure(figsize=(8, 6))
frequency_table.plot(kind='bar', color=['blue', 'orange'], alpha=0.7)
plt.title("Biểu đồ tần số của GT (Giới Tính)", fontsize=14)
plt.xlabel("Giới Tính", fontsize=12)
plt.ylabel("Tần số", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("frequency_bar_chart_GT.png")
plt.show()

plt.figure(figsize=(8, 6))
relative_frequency_table.plot(
    kind='pie', autopct='%1.1f%%', startangle=90, colors=['blue', 'orange'])
plt.title("Biểu đồ tần suất của GT (Giới Tính)", fontsize=14)
plt.ylabel("")
plt.tight_layout()
plt.savefig("relative_frequency_pie_chart_GT.png")
plt.show()
```

# 2. Hãy trình bày dữ liệu lần lượt các biến: US_TBM1, US_TBM2 và US_TBM3

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "processed_dulieuxettuyendaihoc.csv"

try:
    df = pd.read_csv(file_path)
    print("Dữ liệu đã được tải thành công!\n")
except FileNotFoundError:
    print(f"Tệp '{file_path}' không tìm thấy. Vui lòng kiểm tra đường dẫn.")
    exit()

columns = ["US_TBM1", "US_TBM2", "US_TBM3"]
df = df[columns].dropna()

print("🔹 Thống kê dữ liệu của US_TBM1, US_TBM2, US_TBM3:")
print(df.describe())

plt.figure(figsize=(12, 5))
for i, col in enumerate(columns):
    plt.subplot(1, 3, i+1)
    sns.histplot(df[col], bins=15, kde=True, color='blue', alpha=0.7)
    plt.title(f'Phân bố {col}')
    plt.xlabel("Điểm số")
    plt.ylabel("Tần số")

plt.tight_layout()
plt.savefig("histogram_US_TBM.png")
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(data=df, palette="Set2")
plt.title("Boxplot của US_TBM1, US_TBM2 và US_TBM3", fontsize=14)
plt.xlabel("Môn học")
plt.ylabel("Điểm số")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.savefig("boxplot_US_TBM.png")
plt.show()
```

# 3. Hãy trình bày dữ liệu biến DT với các học sinh là nam

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "processed_dulieuxettuyendaihoc.csv"
try:
    df = pd.read_csv(file_path)
    print("Dữ liệu đã được tải thành công!")
except FileNotFoundError:
    print(f"Tệp '{file_path}' không tìm thấy. Vui lòng kiểm tra đường dẫn.")
    exit()
print("Các giá trị trong cột GT:", df["GT"].unique())

df_nam = df[df["GT"].str.strip() == "M"]

print(f"Số lượng học sinh Nam: {len(df_nam)}")

df_nam["DT"] = pd.to_numeric(df_nam["DT"], errors="coerce")

df_nam = df_nam.dropna(subset=["DT"])

frequency_table = df_nam["DT"].value_counts()
relative_frequency_table = df_nam["DT"].value_counts(normalize=True) * 100

print("\nBảng tần số của biến DT:")
print(frequency_table)
print("\nBảng tần suất của biến DT (%):")
print(relative_frequency_table)

plt.figure(figsize=(8, 6))
frequency_table.plot(kind='bar', color='blue', alpha=0.7)
plt.title("Biểu đồ tần số của DT (dành cho học sinh Nam)", fontsize=14)
plt.xlabel("DT", fontsize=12)
plt.ylabel("Tần số", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("frequency_bar_chart_DT_Nam.png")
plt.show()

plt.figure(figsize=(8, 6))
relative_frequency_table.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap="Blues")
plt.title("Biểu đồ tần suất của DT (dành cho học sinh Nam)", fontsize=14)
plt.ylabel("")
plt.tight_layout()
plt.savefig("relative_frequency_pie_chart_DT_Nam.png")
plt.show()
```

# 4. Hãy trình bày dữ liệu biến KV với các học sinh là nam thuộc dân tộc Kinh, có điểm thỏa mãn điều kiện (DH1 >= 5.0 và DH2 >= 4.0 và DH3 >= 4.0)


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "processed_dulieuxettuyendaihoc.csv"
try:
    df = pd.read_csv(file_path)
    print("Dữ liệu đã được tải thành công!")
except FileNotFoundError:
    print(f"Tệp '{file_path}' không tìm thấy. Vui lòng kiểm tra đường dẫn.")
    exit()
print("Các giá trị trong cột GT:", df["GT"].unique())
print("Các giá trị trong cột DT:", df["DT"].unique())

df["DT_encoded"] = df["DT"].apply(lambda x: 0 if x == 0 else 1)

df_filtered = df[(df["GT"].str.strip() == "M") & (df["DT_encoded"] == 0)]

df_filtered["DH1"] = pd.to_numeric(df_filtered["DH1"], errors="coerce")
df_filtered["DH2"] = pd.to_numeric(df_filtered["DH2"], errors="coerce")
df_filtered["DH3"] = pd.to_numeric(df_filtered["DH3"], errors="coerce")

df_filtered = df_filtered[(df_filtered["DH1"] >= 5.0) & 
                          (df_filtered["DH2"] >= 4.0) & 
                          (df_filtered["DH3"] >= 4.0)]

print(f"Số lượng học sinh Nam thuộc dân tộc Kinh đạt điều kiện: {len(df_filtered)}")

frequency_table = df_filtered["KV"].value_counts()
relative_frequency_table = df_filtered["KV"].value_counts(normalize=True) * 100

print("\nBảng tần số của biến KV:")
print(frequency_table)
print("\nBảng tần suất của biến KV (%):")
print(relative_frequency_table)

plt.figure(figsize=(8, 6))
frequency_table.plot(kind='bar', color='blue', alpha=0.7)
plt.title("Biểu đồ tần số của KV (học sinh Nam dân tộc Kinh, đủ điều kiện)", fontsize=14)
plt.xlabel("Khu vực (KV)", fontsize=12)
plt.ylabel("Tần số", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("frequency_bar_chart_KV.png")
plt.show()

plt.figure(figsize=(8, 6))
relative_frequency_table.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap="Blues")
plt.title("Biểu đồ tần suất của KV (học sinh Nam dân tộc Kinh, đủ điều kiện)", fontsize=14)
plt.ylabel("")
plt.tight_layout()
plt.savefig("relative_frequency_pie_chart_KV.png")
plt.show()
```

# 5. Hãy trình bày dữ liệu lần lượt các biến DH1, DH2, DH3 lớn hơn bằng 5.0 và thuộc khu vực 2NT

```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "processed_dulieuxettuyendaihoc.csv"
df = pd.read_csv(file_path)

print("Các giá trị trong cột KV:", df["KV"].unique())

df_filtered = df[(df["DH1"] >= 5.0) & (df["DH2"] >= 5.0) & (df["DH3"] >= 5.0) & (df["KV"] == "2NT")]

print("Số lượng học sinh thỏa mãn điều kiện:", len(df_filtered))
print(df_filtered[["DH1", "DH2", "DH3", "KV"]])

plt.figure(figsize=(8, 6))
df_filtered[["DH1", "DH2", "DH3"]].mean().plot(kind="bar", color=["blue", "green", "red"], alpha=0.7)
plt.title("Điểm trung bình các môn của học sinh KV 2NT (>= 5.0)", fontsize=14)
plt.xlabel("Môn học", fontsize=12)
plt.ylabel("Điểm trung bình", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
df_filtered[["DH1", "DH2", "DH3"]].mean().plot(kind="pie", autopct="%1.1f%%", startangle=90, colors=["blue", "green", "red"])
plt.title("Tỷ lệ điểm trung bình các môn học của học sinh KV 2NT (>= 5.0)", fontsize=14)
plt.ylabel("")
plt.tight_layout()
plt.show()
```
