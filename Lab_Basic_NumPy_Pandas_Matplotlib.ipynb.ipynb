# NUMPY
# 1.1 Tạo mảng với NUMPY


```python
import numpy as np
# Tạo mảng 1 Chiều
mang_1_chieu = np.array([1,3,5,7,9])
print(mang_1_chieu)
# Tạo mảng 2 Chiều
mang_2_chieu = np.array([[1,3,5,7,9],[2,4,6,8,10]])
print(mang_2_chieu)
#Tạo bảng ngẫu nhiên
mang_ngau_nhien = np.random.rand(3,3)
print(mang_ngau_nhien)
```

# Các thao tác cơ bản của numpy

```python
mang_1_chieu = np.array([1,3,5,7,9])

#Cộng trừ nhân chia với một số
sum = mang_1_chieu + 5
multiple = mang_1_chieu * 10

print(sum)
print(multiple)
#Tính tổng, trung bình, lớn nhất, nhỏ nhất
print(mang_1_chieu.sum())
print(mang_1_chieu.mean())
print(mang_1_chieu.max())
print(mang_1_chieu.min())
#Lấy phần tử bất kỳ trong bảng
print(mang_1_chieu[1])
```

# PANDAS

```python
import pandas as pd

data = {
    "Tên": ["Anh", "Đạt", "Trâm", "Linh"],
    "Tuổi": [30, 50, 70, 98],
    "Điểm": [55, 85, 100, 46]
}
df = pd.DataFrame(data)
print(df)
```

```python
# In ra cột tuổi
print(df["Tuổi"])
# In ra hàng đầu tiên
print(df.iloc[0])
# In ra dữ liệu có điểm lớn hơn 80
print(df[df["Điểm"]>80])
# Tính trung bình điểm lớp học
print(df["Điểm"].mean())
```

# MATPLOTLIB

```python
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ đường
x = [1,3,5,7,8]
y = [3,9,87,5,4]

plt.plot(x,y, label="Biểu đồ đường", color="blue", marker="o")
plt.title("Biểu đồ đường")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.legend()
plt.show()
```

```python
# Biểu đồ cột
Ten = ["Ly", "Mai", "Xuan", "Bich", "An", "Khang", "Huy"]
Diem = [5,7,3,5,2,7,10]

plt.bar(Ten,Diem, color="red")
plt.title("Biểu đồ cột")
plt.xlabel("Tên")
plt.ylabel("Điểm")
plt.show()
```

```python
#Biểu đồ phân tán
x = np.random.rand(100)
y = np.random.rand(100)
size = np.random.rand(100) * 100

plt.scatter(x,y, size, alpha = 0.5, c='purple')
plt.title("Biểu đồ phân tán")
plt.xlabel("Trục x")
plt.ylabel("Trục y")
plt.show()
```
