# Kiểm Tra Giữa Kì

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### 1.	Tạo Dataframe như dữ liệu đã cho trên. 

```python
data = {
    'Name': ['Mr.Harris', 'Mrs.Bradley', 'Miss.Laina', 'Mrs.Heath', 'Mr.William', 'Mr.James', 
             'Mr.Timothy', 'Mrs.Oscar', 'Mrs.Nicholas', 'Miss.Marguerite', 'Mr.William', 
             'Miss.Hulda', 'Mr.Lawrence', 'Mr.Wilam'],
    'Height': [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    'Weight': [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    'AGE': [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}
df = pd.DataFrame(data)
```

### 2.	Hiển thị thông tin Dataframe vừa tạo. 

```python
print("Thông tin DataFrame:")
print(df.info())
print("\nDữ liệu DataFrame:")
print(df)
```

### 3.	Tính giá trị trung bình của cột AGE.

```python
age_mean = df['AGE'].mean()
print(f"\nGiá trị trung bình của cột AGE: {age_mean}")
```

### 4.	Nhóm dữ liệu cột Height và tính trung bình cột Weight.

```python
grouped = df.groupby('Height')['Weight'].mean()
print("Trung bình Weight theo Height:")
print(grouped)
```

### 5.	Sắp xếp Dataframe theo cột Weight tăng dần. 

```python
df_sorted = df.sort_values(by='Weight')
print("\nDataFrame sắp xếp theo Weight tăng dần:")
print(df_sorted)
```

### 6. Vẽ biểu đồ cột cho AGE. 

```python
plt.figure(figsize=(14, 6))
plt.bar(df['Name'], df['AGE'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.title('Biểu đồ cột AGE')
plt.xlabel('Name')
plt.ylabel('AGE')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 7.	Vẽ biểu đồ phân tán cho AGE và Height. 

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['AGE'], df['Height'], color='orange', s=100)
plt.title('Biểu đồ phân tán AGE và Height')
plt.xlabel('AGE')
plt.ylabel('Height (m)')
plt.grid(True, linestyle='--', alpha=0.7)
for i, name in enumerate(df['Name']):
    plt.annotate(name, (df['AGE'][i], df['Height'][i]), fontsize=8, ha='right')
plt.tight_layout()
plt.show()
```

### 8.	Vẽ biểu đồ đường cho cột Weight. 

```python
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Weight'], marker='o', color='green', linestyle='-', linewidth=2)
plt.xticks(df.index, df['Name'], rotation=45, ha='right')
plt.title('Biểu đồ đường Weight')
plt.xlabel('Người')
plt.ylabel('Weight (kg)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### 9.	Vẽ biểu đồ tròn cho cột Height. 

```python
plt.figure(figsize=(8, 8))
plt.pie(df['Height'], labels=df['Name'], autopct='%1.1f%%')
plt.title('Biểu đồ tròn Height')
plt.show()
```

### 10.	Vẽ biểu đồ phân tán cho cột Weight,AGE. 

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Weight'], df['AGE'], color='purple', s=100)
plt.title('Biểu đồ phân tán Weight và AGE')
plt.xlabel('Weight (kg)')
plt.ylabel('AGE')
plt.grid(True, linestyle='--', alpha=0.7)
for i, name in enumerate(df['Name']):
    plt.annotate(name, (df['Weight'][i], df['AGE'][i]), fontsize=8, ha='right')
plt.tight_layout()
plt.show()
```

### 11.	Kiểm tra có giá trị NaN nào trong DataFrame không.

```python
print("\nKiểm tra giá trị NaN:")
print(df.isna().sum())
```

### 12.	Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên. 

```python
df_unique = df.drop_duplicates(subset='Name', keep='first')
print("\nDataFrame sau khi loại bỏ trùng lặp:")
print(df_unique)
```

### 13.	Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó .

```python
df_unique.loc[df_unique['AGE'] > 40, 'AGE'] = age_mean
print("\nDataFrame sau khi thay AGE > 40 bằng trung bình:")
print(df_unique)
```

### 14.	Biết công thức tính chỉ số BMI như sau: 
### BMI = Weight(kg)/Height2 (m2)  
### Tạo một cột mới trong Dataframe với tên <cs_BMI>, tính chỉ số BMI mỗi người 


```python
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

print("\nDataFrame với cột cs_BMI:")
print(df)

print("\nKiểm tra từng giá trị BMI:")
for i, row in df.iterrows():
    bmi = row['Weight'] / (row['Height'] ** 2)
    print(f"{row['Name']}: BMI = {bmi:.2f}")
```

### 15.	Theo thang đo kết quả tính BMI của châu Á: 
### Chỉ số BMI dưới 18,5 là thiếu cân, 18.5-22,9 là bình  	thường, 23-24,9 là thừa cân, từ 25 trở lên là béo phì. 
### Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng với mức chỉ số BMI trên.


```python
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0
    elif 18.5 <= bmi <= 22.9:
        return 1
    elif 23 <= bmi <= 24.9:
        return 2
    else:
        return 3

df['Phan_Loai'] = df['cs_BMI'].apply(classify_bmi)

print("\nDataFrame với cột Phan_Loai:")
print(df)

print("\nKiểm tra phân loại BMI (Phan_Loai):")
for i, row in df.iterrows():
    bmi = row['cs_BMI']
    phan_loai = classify_bmi(bmi)
    print(f"{row['Name']}: BMI = {bmi:.2f}, Phan_Loai = {phan_loai}")
```

### 16.	Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1. 

```python
df['cs_BMI_normalized'] = (df['cs_BMI'] - df['cs_BMI'].min()) / (df['cs_BMI'].max() - df['cs_BMI'].min())
print("\nDataFrame với cột cs_BMI_normalized:")
print(df.round(4))
print("\nKiểm tra từng giá trị cs_BMI_normalized:")
min_bmi = df['cs_BMI'].min()
max_bmi = df['cs_BMI'].max()
for i, row in df.iterrows():
    bmi = row['cs_BMI']
    normalized = (bmi - min_bmi) / (max_bmi - min_bmi)
    print(f"{row['Name']}: cs_BMI = {bmi:.2f}, cs_BMI_normalized = {normalized:.4f}")
```

### 17.	Cho biết: 
###  + ‘Mr‘tương ứng với ‘male’, 
###  + ’Mrs’ và ‘Miss’ tương ứng với ‘female’.  
### tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó. 


```python
def get_gender(name):
    if 'Mr.' in name:
        return 'male'
    elif 'Mrs.' in name or 'Miss.' in name:
        return 'female'
    return None

df_unique['GT'] = df_unique['Name'].apply(get_gender)
df_unique['Name'] = df_unique['Name'].str.replace('Mr. ', '').str.replace('Mrs. ', '').str.replace('Miss. ', '')
print("\nDataFrame với cột GT và Name đã xóa tiền tố:")
print(df_unique)
```

### 18.	Vẽ biểu đồ tròn cho GT. 

```python
plt.figure(figsize=(8, 8))
gt_counts = df_unique['GT'].value_counts()
plt.pie(gt_counts, labels=gt_counts.index, autopct='%1.1f%%', colors=['lightblue', 'pink'])
plt.title('Biểu đồ tròn Giới tính (GT)')
plt.show()
```

### 19.	Trực quan hóa dữ liệu người nữ trên <Phan_Loai>. 

```python
df_unique['cs_BMI'] = df_unique['Weight'] / (df_unique['Height'] ** 2)
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0
    elif 18.5 <= bmi <= 22.9:
        return 1
    elif 23 <= bmi <= 24.9:
        return 2
    else:
        return 3
df_unique['Phan_Loai'] = df_unique['cs_BMI'].apply(classify_bmi)
female_data = df_unique[df_unique['GT'] == 'female']
phan_loai_counts = female_data['Phan_Loai'].value_counts()
plt.figure(figsize=(10, 6))
plt.bar(phan_loai_counts.index, phan_loai_counts.values, color='pink')
plt.title('Phân loại BMI của người nữ')
plt.xlabel('Phan_Loai (0: Thiếu cân, 1: Bình thường, 2: Thừa cân, 3: Béo phì)')
plt.ylabel('Số lượng')
plt.xticks([0, 1, 2, 3])
plt.show()
```
